```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import lasio
```

```python
df =pd.read_csv('ID1_Data.csv')
```

```python
df
```

```python
GR = df['GR'].to_numpy()

GR_min = np.percentile(GR,5)
GR_max = np.percentile(GR,95)

IGR = (GR - GR_min)/(GR_max - GR_min)

IGR
```

```python
Vsh = 0.33*(2**(2*IGR) - 1)
Vsh
```

```python
plt.figure(figsize = (4,8))
plt.plot(df['GR'],df['DEPTH'],label = 'GR')
plt.plot(np.ones(len(GR))*GR_min,df['DEPTH'],label = 'Sand line')
plt.plot(np.ones(len(GR))*GR_max,df['DEPTH'],label = 'Shale line')
plt.legend(loc='upper right')
plt.xlabel('GR')
plt.ylabel('Depth')
```

```python
df['GR'].plot(kind='hist', bins=40, density=True,edgecolor='black')
df['GR'].plot(kind='kde',color='black')
plt.xlabel('GR[API]', fontsize=14)
mean=df['GR'].mean()
p05=df.GR.quantile(0.05)
p95=df.GR.quantile(0.95)
plt.axvline(mean,color='g',label='mean')
plt.axvline(p05,color='r',label='p5')
plt.axvline(p95,color='y',label='p95')
plt.legend()
plt.xlim(0,200)
```
