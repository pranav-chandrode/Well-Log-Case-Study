```python
def plot_graphs(df):
    fig,ax = plt.subplots(nrows=2,ncols=6,figsize=(12,12))
    color_lst = ['r','g','b','y','m']
    bit_size = 8.5
    
    idx  =int(0)
    for i in range(2):
        for j in range(6):
            if i == 0 and j==0:
                ax[i][j].plot(df[df.columns[(idx+1)%(len(df.columns))]],df['DEPTH'],color = color_lst[idx%len(color_lst)],linewidth = 1.5)
                ax[i][j].axvline(bit_size)
                    
            ax[i][j].plot(df[df.columns[(idx+1)%(len(df.columns))]],df['DEPTH'],color = color_lst[idx%len(color_lst)],linewidth = 1.5)
            ax[i][j].grid()
            ax[i][j].invert_yaxis()
            ax[i][j].set_title(df.columns[(idx+1)%(len(df.columns))])
            if i!=0:
                ax[i][j].set_yticklabels([])
            idx +=1
```

```python
plot_graphs(df)
```

```python
def plot_hist(df):
    color_lst = ['r','g','b','y','m']
    logs =  ['GR','RHOM','APLU']
    fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(10,4))
    for i in range(3):
        sns.histplot(df[logs[i]],kde =True, ax = axes[i],color=color_lst[i])
        
```

```python
plot_hist(df)
```

### Volume of Shale

```python
GR = df['GR'].to_numpy()

GR_min = np.percentile(GR,5)
GR_max = np.percentile(GR,95)

IGR = (GR - GR_min)/(GR_max - GR_min)

# Vsh = 0.33*(2**(2*IGR) - 1)
Vsh = IGR

plt.figure(figsize = (4,8))
plt.plot(df['GR'],df['DEPTH'],label = 'GR')
plt.plot(np.ones(len(GR))*GR_min,df['DEPTH'],label = 'Sand line')
plt.plot(np.ones(len(GR))*GR_max,df['DEPTH'],label = 'Shale line')
plt.legend(loc='upper right')
plt.xlabel('GR')
plt.ylabel('Depth')
```
