```python
import lasio
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
las=lasio.read(r"IchthysDeep-1_client_edit.las")
```

```python
df=las.df().reset_index()
df.head()
```

```python
df=df[(df["DEPT"]>=4074.5) & (df["DEPT"]<=4242.9)]

NPHI= df['APLC'] + 0.04

por_d = (2.65 - df['RHOM'])/(2.65 - 1.1)

total_por= (NPHI + por_d)/2
# total_por= ((NPHI**2 + por_d**2)/2)**0.5
```

```python
fig ,ax = plt.subplots(1,2,figsize = (6,10))
logs = ['AFRT','AFRX']
color_lst = ['c','g','b','y','m']

for i in range(2):
    ax[i].plot(np.log10(df[logs[i]]),df['DEPT'],color = color_lst[i],linewidth = 1)
    ax[i].set_title(logs[i])
    ax[i].invert_yaxis()

ax[0].set_ylabel('Depth (m)')
for i in range(2):
    if i!=0:
        ax[i].set_yticklabels([])
```

```python
depth = df['DEPT']

fig , ax = plt.subplots(1,2,figsize = (6,10))

ax11 = ax[0].twiny()
ax[0].invert_yaxis()
# ax11.invert_yaxis()
ax11.grid(True)
# ax21.set_xlim(140,40)
ax11.spines['top'].set_position(('outward', 0))
ax11.plot(NPHI, depth, '-b', linewidth = 0.8)
ax11.set_xlabel('Neutron Porosity', color= 'b')
ax11.tick_params(axis='x', colors= 'b')


ax12 = ax[0].twiny()
# ax22.set_xlim(0, 60)
# ax12.invert_yaxis()
# ax12.invert_xaxis()
ax12.spines['top'].set_position(('outward', 40))
ax12.plot(por_d, depth, '--k',linewidth = 1)
ax12.set_xlabel('Density Porosity', color = 'k')
ax12.tick_params(axis= 'x', colors='k')


ax21 = ax[1].twiny()
ax[1].invert_yaxis()
# ax21.invert_yaxis()
ax21.grid(True)
# ax21.set_xlim(140,40)
ax21.spines['top'].set_position(('outward', 0))
ax21.plot(total_por, depth, '-b', linewidth = 0.8)
ax21.set_xlabel('Total Porosity', color= 'b')
ax21.tick_params(axis='x', colors= 'b')
```

```python
las2=lasio.read('ichthys_deep_1_wire_public_2010_sdb.las')
df2=las2.df().reset_index()
df2.head()
```

```python
data1 = df1[(df1["DEPT"]>=4074.5) & (df1["DEPT"]<=4242.9)]
data2 = df2[(df2["DEPTH"]>=4074.5) & (df2["DEPTH"]<=4242.9)]


GR = data1['GR'].to_numpy()

GR_min = np.percentile(GR,5)
GR_max = np.percentile(GR,95)

IGR = (GR - GR_min)/(GR_max - GR_min)

# Vsh = 0.33*(2**(2*IGR) - 1)
Vsh = IGR
```
